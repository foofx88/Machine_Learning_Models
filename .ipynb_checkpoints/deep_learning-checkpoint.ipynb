```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]
X = df.drop("koi_disposition", axis=1)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df["koi_disposition"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

# Label-encode y
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

# Converting to One-Hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

# Train the Model



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create model and add layers
deep_model = Sequential()
deep_model.add(Dense(units=50, activation='relu', input_dim=40)) #initial layer with 40 columns
deep_model.add(Dense(units=50, activation='relu')) # 1st hidden layer
deep_model.add(Dense(units=3, activation='softmax')) #Outputs- Confirmed, Candidate, False Positive
# Compile and fit the model
deep_model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# model.summary()

deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=150,
    shuffle=True,
    verbose=2)
```

# Checking the Model


```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss:.3f}, Accuracy: {model_accuracy:.3f}")
```

```python
#Making prediciotns with the model

encoded_prediction = deep_model.predict_classes(X_test_scaled[:10])
prediction_label = label_encoder.inverse_transform(encoded_prediction)
print(f"Predicted classes: {list(prediction_label)}")
print(f"Actual Labels: {list(y_test[:10])}")
```
